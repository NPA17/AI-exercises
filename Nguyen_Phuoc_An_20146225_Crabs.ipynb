```python
import pandas as pd
url = 'crabs.csv'
data = pd.read_csv(url)
dataframe = pd.read_csv(url)
data.head()
list(dataframe.columns)
df = dataframe

data = data.replace('B',0)
data = data.replace('O',1)
data = data.replace('M',1)
data = data.replace('F',0)

df.drop('sex',axis = 1, inplace = True)
print(dataframe)

from sklearn import datasets
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

x = data.drop(['sex'], axis =1)
y = data['sex']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
sc=StandardScaler()
sc.fit(x_train)
x_train_std = sc.transform(x_train)
x_test_std = sc.transform(x_test)
print(x_train,x_train_std)





model = Perceptron(max_iter = 1000, eta0= 0.5, random_state = 0)
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
print(y_pred)
print(y_test)
print('Do chinh xac mo hinh chua Scale: ', accuracy_score(y_test,y_pred))
#Mo hinh Scale
model = Perceptron(max_iter= 1000, eta0= 0.5, random_state= 0)
model.fit(x_train_std,y_train)
y_pred = model.predict(x_test_std)
print(y_pred)
print(y_test)
print('Do chinh xac mo hinh da Scale:',accuracy_score(y_test,y_pred))
```
